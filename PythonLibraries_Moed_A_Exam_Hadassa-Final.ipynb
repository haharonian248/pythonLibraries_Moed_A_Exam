```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Import data:

items_df = pd.read_csv("C:/AI Course/Items.csv")
customers_df = pd.read_csv("C:/AI Course/customers.csv")
orders_df = pd.read_csv("C:/AI Course/orders.csv")
order_item_df = pd.read_csv("C:/AI Course/order_item.csv")
```

```python
# Copy of the data:

items_df_copy = items_df.copy()
customers_df_copy = customers_df.copy()
orders_df_copy = orders_df.copy()
order_item_df_copy = order_item_df.copy()
```

```python
# Data preparation:
# Exe1- a- Remove duplicate items. Condition 1: Same item name and item brand:

duplicate_items = items_df_copy[items_df_copy.duplicated(['item_name','item_brand'], keep=False)]
duplicate_items
print('No duplicates found with first condition')
```

```python
# 2nd condition: Same item name, item brand with value and item brand without value:
item_name_grouping = items_df_copy.groupby(['item_name','item_brand'], dropna=False).size()
item_name_grouping = item_name_grouping.reset_index(name='count')
item_name_grouping
print('Found duplicates in Headphones with second condition')
```

```python
# Change id= 8 to 49:
item_ids_map = {8:49}
items_df_copy.set_index('id', inplace=True)
```

```python
# b- Removes the duplicate headphone id=8 as it has less data:
items_df_copy.drop(item_ids_map.keys(), axis=0, inplace=True)
```

```python
# reset index in table: items_df_copy:
items_df_copy.reset_index('id', inplace=True)
items_df_copy
```

```python
# Replace item ID=49 to 8 in the order_item table:
order_item_df_copy['item_id'] = order_item_df_copy['item_id'].replace(item_ids_map)
order_item_df_copy
```

```python
# Remove irrelevant column in order_item_df_copy:
order_item_df_copy.drop(['Unnamed: 0'], axis=1, inplace=True)
```

```python
# Exe2- Handle duplicate data in customer dataset:
# a- Duplicates are customers with the same email address:

duplicate_customers = customers_df_copy[customers_df_copy.duplicated('email', keep=False)]
print('No duplicates found in the customers dataset.')
```

```python
# Exe 3- Handle missing data in all datasets:
# Items_df_copy- missing data:
missing_data = items_df_copy.isnull().sum()
total_rows = items_df_copy.isnull().count()
percentage_missing = missing_data * 100 / total_rows
final_missing_data = pd.concat([total_rows, missing_data, percentage_missing], axis=1, keys=['Total rows', 'Missing rows', 'Percentage'])
final_missing_data
```

```python
## Filling missing data in items_df_copy with default values:
# item_category- above 5% so we will add default value of miscat as we don't know the category it belongs to:
items_df_copy['item_category'] = items_df_copy['item_category'].fillna('Miscat')
# stock_quantity is an important column, without it we can't know how many are left so we will need to remove these rows:
items_df_copy.dropna(subset=['stock_quantity'], inplace=True)
# Item_brand is under 5% so we will remove these rows:
items_df_copy.dropna(subset=['item_brand'], inplace=True)
items_df_copy
```

```python
# Drop rows from order_item_df_copy that are pointing to the missing items
order_item_df_copy = order_item_df_copy[order_item_df_copy['item_id'].isin(items_df_copy['id'])]
```

```python
# customers_df_copy- missing data:
missing_data = customers_df_copy.isnull().sum()
total_rows = customers_df_copy.isnull().count()
percentage_missing = missing_data * 100 / total_rows
final_missing_data = pd.concat([total_rows, missing_data, percentage_missing], axis=1, keys=['Total rows', 'Missing rows', 'Percentage'])
final_missing_data
```

```python
## Filling missing data with default values:
# Gender is not mandatory but over 5% so we will use "Unknown":
customers_df_copy['gender'] = customers_df_copy['gender'].fillna('Unknown')
# Age is not mandatory but over 5% so will use average age:
customers_df_copy['age'] = customers_df_copy['age'].fillna(round(customers_df_copy['age'].mean()))
# Joining date is not mandatory but over 5% so we will use a default date so that we can filter them out if necessary. The default date is out of the range so it will be easy to filter:
customers_df_copy['joining_date'] = customers_df_copy['joining_date'].fillna('1999-01-01')
```

```python
# Orders_df_copy- missing data:
missing_data = orders_df_copy.isnull().sum()
total_rows = orders_df_copy.isnull().count()
percentage_missing = missing_data * 100 / total_rows
final_missing_data = pd.concat([total_rows, missing_data, percentage_missing], axis=1, keys=['Total rows', 'Missing rows', 'Percentage'])
final_missing_data
```

```python
## Filling missing data in orders_df_copy with default values:
# Customer_id is over 5% but it is a mandatory field so we will remove these rows:
orders_df_copy.dropna(subset=['customer_id'], inplace=True)
```

```python
# Delivery days isn't a mandatory field but it's above 5% so we will put a default value of average delivery days:
orders_df_copy['delivery_days'] = orders_df_copy['delivery_days'].fillna(round(orders_df_copy['delivery_days'].mean()))
```

```python
# Drop rows from order_item_df_copy that are pointing to the missing orders as we removed them because they were missing the customer ids:
order_item_df_copy = order_item_df_copy[order_item_df_copy['order_id'].isin(orders_df_copy['order_id'])]
```

```python
# order_item_df_copy- missing data:
missing_data = order_item_df_copy.isnull().sum()
total_rows = order_item_df_copy.isnull().count()
percentage_missing = missing_data * 100 / total_rows
final_missing_data = pd.concat([total_rows, missing_data, percentage_missing], axis=1, keys=['Total rows', 'Missing rows', 'Percentage'])
final_missing_data
```

```python
# Data Analysis:

#Exe 1- Explore customers_df_copy:
# a- Create dataframe:
gender_customers_df_copy = customers_df_copy.groupby('gender').size()
gender_customers_df_copy = gender_customers_df_copy.reset_index(name='count')
gender_customers_df_copy
```

```python
# a- Create bar chart:
plt.figure(figsize=(12,6))
plt.bar(gender_customers_df_copy['gender'], gender_customers_df_copy['count'])
plt.title('Number of customers by gender')
plt.xlabel('Gender')
plt.ylabel('Count of customers')
plt.show()
```

```python
# b- Create a histogram of the customer age distribution:
plt.figure(figsize=(12,6))
plt.hist(customers_df_copy['age'], bins=20)
plt.title('Age distribution of customers')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
# c- What is the year with the most joining customers:
customers_df_copy['joining_date'] = pd.to_datetime(customers_df_copy['joining_date'])
customers_df_copy.set_index('joining_date', inplace=True)
df_resample = customers_df_copy.resample(rule='YE')['id'].size()
df_resample = df_resample.reset_index(name='count').sort_values(by='count', ascending=False)
print(f"The year with the most joining customers to the company is: {df_resample.iloc[0]['joining_date'].year}, with {df_resample.iloc[0]['count']} customers.")
```

```python
# d- Create a line chart which will show if there is a pattern in the number of joining customers over the years 2020- 2023:
# First I will filter out only the years 2020- 2023 as I added a default year and it adds in the dataframe all the years till 2020
df_filtered = df_resample[(df_resample['joining_date'].dt.year >= 2019) & (df_resample['joining_date'].dt.year <= 2023)]
df_filtered = df_filtered.sort_values(by='joining_date')
plt.figure(figsize=(12,6))
plt.plot(df_filtered['joining_date'], df_filtered['count'], marker='o')
plt.title('Trend of number of joining customers between 2020-2023')
plt.xlabel('Year')
plt.ylabel('Number of joining customers')
plt.grid()
plt.show()
print("According to the graph it looks like the amount of customers joining is going up.")
```

```python
# e- Check if there is a particular month that consistently sees a higher number of new customers.
df_resample_month = customers_df_copy.resample(rule='ME')['id'].size()
df_resample_month = df_resample_month.reset_index(name='count')
df_filtered = df_resample_month[(df_resample_month['joining_date'].dt.year >= 2020) & (df_resample_month['joining_date'].dt.year <= 2023)]
plt.figure(figsize=(15,6))
plt.plot(df_filtered['joining_date'], df_filtered['count'], marker='o')
plt.title('Trend of number of joining customers between 2020-2023')
plt.xlabel('Month')
plt.ylabel('Number of joining customers')
plt.grid()
plt.show()
print("There are a few peaks in the graph but they aren't consistent. From 2021, in september there is a peak which continues also in 2022 but there isn't enther data to show it's consistent.")
```

```python
# Exe2 Explore items_df_copy:
# a- Create a dataframe showing the number of items in each categpry:
item_category_grouping = items_df_copy.groupby('item_category')['id'].count()
item_category_grouping = item_category_grouping.reset_index(name='count')
item_category_grouping
```

```python
plt.figure(figsize=(15,10))
plt.xticks(rotation=90)
plt.bar(item_category_grouping['item_category'], item_category_grouping['count'])
plt.title('Number of items per category')
plt.xlabel('Category')
plt.ylabel('Number of items')
plt.show()
```

```python
# b- Item with the largest quantity in stock and the lowest quantity in stock:
max_item_quantity = items_df_copy[items_df_copy['stock_quantity'] == items_df_copy['stock_quantity'].max()]
summary_df = items_df_copy[
    items_df_copy['stock_quantity'].isin([
        items_df_copy['stock_quantity'].max(),
        items_df_copy['stock_quantity'].min()
    ])
]
summary_df
```

```python
# c- The average stock quantity of all items:
average_stock_quantity = round(items_df_copy['stock_quantity'].mean())
print(f"The average stock quantity of all the items is: {average_stock_quantity}")
```

```python
# Exe3- Explore orders_df_copy
# a- Create a data frame with the customer purchase amount
customer_purchase_amount = orders_df_copy.groupby('customer_id')['order_id'].count()
customer_purchase_amount = customer_purchase_amount.reset_index(name='count_orders')
customer_purchase_amount_grouping = customer_purchase_amount.groupby('count_orders').size()
customer_purchase_amount_grouping = customer_purchase_amount_grouping.reset_index(name = 'amount_of_customers')
customer_purchase_amount_grouping
```

```python
# Create a bar chart for visualization:
plt.bar(customer_purchase_amount_grouping['count_orders'], customer_purchase_amount_grouping['amount_of_customers'])
plt.title('Amount of customers according to the amount of orders')
plt.xlabel('Amount of orders')
plt.ylabel('Amount of customers')
plt.show()
```

```python
# b- The top 5 customers with the highest amount of purchases:
customer_purchase_amount = customer_purchase_amount.sort_values(by='count_orders', ascending=False).head(5)
selected_columns = ['customer_id','first_name','last_name','count_orders']
customer_purchase_amount_details = customers_df_copy.merge(customer_purchase_amount, left_on='id', right_on='customer_id', how='inner')[selected_columns]
customer_purchase_amount_details = customer_purchase_amount_details.sort_values(by='count_orders', ascending=False)
customer_purchase_amount_details
```

```python
#c- How many customers are paying by each payment method:
customer_purchase_method_amounts = orders_df_copy.groupby('payment_method')['customer_id'].count()
customer_purchase_method_amounts = customer_purchase_method_amounts.reset_index(name='amount_of_customers')
customer_purchase_method_amounts
```

```python
# Create a bar chart for customer_purchase_method_amounts:
plt.bar(customer_purchase_method_amounts['payment_method'],customer_purchase_method_amounts['amount_of_customers'])
plt.title('Amount of customers according to the payment methods')
plt.xlabel('Payment methods')
plt.ylabel('Amount of customers')
plt.show()
```

```python
# Exe4- Explore all datasets:
# a- Add a new column with the total price per order_id:
order_item_price_df = order_item_df_copy.merge(items_df_copy, left_on='item_id', right_on='id' , how='left') 
order_item_price_df['total_price'] = order_item_price_df['quantity'] * order_item_price_df['item_price']
order_item_price_df = order_item_price_df.groupby('order_id')['total_price'].sum().reset_index(name='total_amount')
totals_map = order_item_price_df.set_index('order_id')['total_amount']
orders_df_copy['total_price'] = orders_df_copy['order_id'].map(totals_map)
```

```python
# b- Show the max, min and average of all the orders total price:
max_order_total_price = orders_df_copy['total_price'].max()
min_order_total_price = orders_df_copy['total_price'].min()
mean_order_total_price = orders_df_copy['total_price'].mean()
print(f"The maximum total price for all the orders is: {max_order_total_price}")
print(f"The minimum total price for all the orders is: {min_order_total_price}")
print(f"The maximum total price for all the orders is: {mean_order_total_price}")
```

```python
# c- Check to see if there is a correlation between the number of items in an order and the delivery time:
total_item_quantity = order_item_df_copy.groupby('order_id')['quantity'].sum()
total_item_quantity = total_item_quantity.reset_index(name='total_item_quantity')
total_quantity_delivery_time = total_item_quantity.merge(orders_df_copy, on='order_id', how='left')
total_quantity_delivery_time
```

```python
# Scatter plot to show if corrolation exists between amount of items and delivery time:
plt.scatter(total_quantity_delivery_time['total_item_quantity'], total_quantity_delivery_time['delivery_days'])
plt.title('Amount of items vs Delivery time')
plt.xlabel('Amount of items')
plt.ylabel('Delivery time')
plt.show()
print("There is no correlation between the amount of items and the delivery time")
```

```python
# d- How many customers never created an order:
customers_orders = customers_df_copy.merge(orders_df_copy, left_on='id', right_on='customer_id', how='left')
customers_no_orders = customers_orders.isnull().sum()
print(f"There are {customers_no_orders.loc['order_id']} customers which have never created an order.")
```

```python
# e- Find the 5 top items which have been ordered the least:
items_quantity = order_item_df_copy.groupby('item_id')['quantity'].sum()
items_quantity = items_quantity.reset_index(name='total_quantity')
items_quantity = items_quantity.sort_values(by='total_quantity').head(5)
items_quantity
```

```python
# Bonus:
# Exe1- The management started in August 2023 a marketing campaign in the USA country to market their website. Was the campain useful or not?
#First, extract only the orders done throught the website:
website_orders_df_copy = orders_df_copy[orders_df_copy['order_source'] == 'WEBSITE']

#Second, extract only the customers from the USA:
usa_customers_df_copy = customers_df_copy[customers_df_copy['nationallity'] == 'United States']
usa_customers_df_copy = usa_customers_df_copy.reset_index()

#Third, combine the tables to get all the data:
website_usa_details = website_orders_df_copy.merge(usa_customers_df_copy, left_on='customer_id', right_on='id', how='inner')
total_amount_items_order_website_usa = website_usa_details.merge(total_item_quantity, on='order_id', how='inner')
total_amount_items_order_website_usa = total_amount_items_order_website_usa.sort_values(by='order_date')
total_amount_items_order_website_usa
```

```python
#Find how many order IDs per order date:
total_amount_items_order_website_usa['order_date'] = pd.to_datetime(total_amount_items_order_website_usa['order_date'])
total_amount_items_order_website_usa.set_index('order_date', inplace=True)
df_resample = total_amount_items_order_website_usa.resample(rule='ME')['order_id'].nunique()
df_resample = df_resample.reset_index(name='count_orders')
df_resample
```

```python
# Line plot
plt.figure(figsize=(15,6))
plt.plot(df_resample['order_date'], df_resample['count_orders'])
plt.title('Amount of USA orders per month done through the website')
plt.xlabel('Order dates- month')
plt.ylabel('Amount of orders')
plt.grid()
plt.show()
print('It looks like the marketing campaign which started in August in the USA to market their website took a bit of time to start helping.\n We see an increase starting from September.')
```

```python
# Exe2: Did the discount price given to those using giftcards increase the sales?
# We will extract only the giftcard payment method from the orders and combine it with the total item quantities:
giftcard_orders_df_copy = orders_df_copy[orders_df_copy['payment_method'] == 'GIFT_CARD']
total_amount_items_order_giftcard = giftcard_orders_df_copy.merge(total_item_quantity, on='order_id', how='inner')
total_amount_items_order_giftcard = total_amount_items_order_giftcard.sort_values(by='order_date')
```
